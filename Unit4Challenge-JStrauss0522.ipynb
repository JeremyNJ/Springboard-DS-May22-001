**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose the df so the dates are the columns ("variables") and the boroughs are the rows
# ("observations")
txp=properties.transpose()
 #txp.shape
 #txp.head()

# Drop the first column, which was the first row, and was some sort of id for the geographic regions.
txp2=txp.drop(0, axis='columns')
print("txp2.index:\n", txp2.index)

# Make the first row (the dates of each observation) the header, and remove it from the data rows
txp2.columns = txp2.iloc[0]
txp2 = txp2[1:]

# Retain only the rows which are boroughs of London. (We also retain City of London, a non-borough
# government district, for now, to see if including it in the analysis may be interesting.)
txp3 = txp2[0:33]
print("\nHere is the transposed dataframe. The first row (the dates) is now column labels.")
print("Only the rows representing the boroughs of London are retained, with the addition")
print("of City of London, a non-borough government entity that may be of interest.")
txp3.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Let's make the names of the boroughs a regular data variable (ie column) and create
# a new column with uniform substrings of the borough names, for easier reference.
# (There is no point making the new column an index, because we will be stacking the data
# to make a long DataFrame, and the boroughs will no longer be unique.)
txp4 = txp3.reset_index()
txp4.rename(columns={"index": "borough"}, inplace=True)
txp4['keyname'] = txp4['borough'].apply(lambda s : s[0:4])
print("col keyname is unique:", txp4['keyname'].is_unique)
#txp4.set_index('keyname', inplace=True)
print("Show renamed borough column.")
print("Note new column at the end with substrings of borough name, for easier reference.")
txp4.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# We will look at the most recent 20 full calendar years, 2002 through 2021. And to look at those
# years, comparing price change over time by borough, we do not need to be concerned with the
# little ups and downs every month. So we will be summarizing each year with an annual average
# by borough, later in this series of data manipulations.
# For now, let's start by stacking the most recent 20 years:
long_df = txp4.melt(id_vars=['borough', 'keyname'], value_vars=txp4.columns[85:325],
                    var_name=['date'], value_name='price')
print("long_df.shape:", long_df.shape)
print("long_df.head():\n", long_df.head())
print("long_df.tail():")
long_df.tail()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
print('dtypes:')
long_df.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
for col in long_df.columns : print("len("+col+"):", len(long_df[col]))
# unique boroughs should be 33 (32 boroughs + City of London)
print("unique boroughs (should be 33 incl City):", len(set(long_df.loc[:,'borough'])))
# Section 2.4, above, revealed unexpected data types. Let's see if the problem is nulls.
for col in long_df.columns : print("isna("+col+").sum:", pd.isna(long_df[col]).sum())
```

```python
# There are no nulls. Let's take a closer look at type for the prices
count = 0
for ind, val in enumerate(long_df['price']) :
    if type(val) == type(1.0) : count += 1
    else : print("price", ind, "type:", type(val))
print("float:", count)
```

```python
# That's easy to fix.
count = 0
fixed = 0
for ind, val in enumerate(long_df['price']) :
    if type(val) == type(1.0) :
        count += 1
    elif type(val) == type(1) :
        print('price at', ind, "was", long_df.loc[long_df.index[ind], 'price'])
        long_df.loc[long_df.index[ind], 'price'] = float(val)
        print('price at', ind, "is now", long_df.loc[long_df.index[ind], 'price'])
        fixed += 1
print("float:", count)
print("fixed:", fixed)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Let's find the 3 boroughs with the highest prices at the end of 2021 and chart the prices
# over time for those 3.
print(long_df[long_df['date']==datetime(2002, 1, 1, 0, 0)].sort_values('price').head(10))
long_df[long_df['date']==datetime(2021, 12, 1, 0, 0)].sort_values('price',
                                                                       ascending=False).head(3)
```

```python
# To take the 3 boroughs with highest prices at the end of 2021 and plot them over time, together,
# we need 3 arrays or one DataFrame with 3 columns. I will create the one DataFrams by subsetting
# long_df on the 3 different boroughs and then merging the results:
df_kens = long_df.loc[:, 'keyname':'price'].query("keyname == 'Kens'")
df_west = long_df.loc[:, 'keyname':'price'].query("keyname == 'West'")
df_camd = long_df.loc[:, 'keyname':'price'].query("keyname == 'Camd'")
df3 = df_kens.merge(df_west, on='date', suffixes=('','_west'))\
             .merge(df_camd, on='date', suffixes=('_kens','_camd'))
print('df3 shape:', df3.shape)
print(df3.head())
df3.plot(x="date", y=["price_kens", "price_west", "price_camd"],
            kind="line", rot=90, ylabel="price",
            label=['Kens & Chelsea','Westminster','Camden'])
```

```python
# Let's get that growth rate. Rather than the suggested ratio of 2021 price over
# 2002 price, we calculate an actual growth rate: growth / original.
# We'll pull out 2002 and 2021 and create a new column for growth.
#
# (NOTE: A more sophisticated analysis of price increases would look at year-over-year
# per-cent increases. These could be compared across boroughs graphically, or boiled down
# to a single statistic: a mean of the 19 year-over-year per-cent increases for each borough.)
#
df_increase = df_years.query("year == 2002 or year == 2021")
print("Before pivot, df_increase.shape:", df_increase.shape)
print("Before pivot, df_increase.head")
print(df_increase.head(), '\n')
df_increase = df_increase.pivot(index="borough", columns="year")
print("After pivot, df_increase.shape", df_increase.shape)
print("After pivot, df_increase.head")
print(df_increase.head())
df_increase.columns = ['2002', '2021']
# create new column
df_increase['growth_rate'] = (df_increase['2021'] - df_increase['2002']) / df_increase['2002']
print("\nLondon boroughs, (plus City of London, an interesting non-borough entity).")
print("Sorted by growth rate. Here are the top 5:")
print("Note that I have calculated an actual growth rate (end - start)/start")
print(df_increase.sort_values('growth_rate', ascending=False).head())
df_increase['end_over_start'] = df_increase['2021'] / df_increase['2002']
print("\nLondon boroughs, (plus City of London, an interesting non-borough entity).")
print("Sorted by the end/start ratio. Here are the top 5:")
print("Note that I have calculated the reciprocal of the brief's suggested start/end ratio,")
print("so that higher numbers denote higher growth.")
df_increase.sort_values('end_over_start', ascending=False).head()
```
